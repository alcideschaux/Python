```python
import pandas as pd
import numpy as np
import scipy.stats
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme(rc={'figure.figsize':[12,8]})
```

# Descripción de variables numéricas

## Medidas de tendencia central

### Promedio (media aritmética)

```python
x = [1,2,3,4,5]
print(x)
```

```python
np.mean(x)
```

```python
x = np.random.normal(size=1000)
np.mean(x)
```

```python
sns.kdeplot(x)
plt.axvline(np.mean(x), color='red')
plt.show()
```

### Mediana

```python
y = [1,2,3,4,100]
print(y)
```

```python
np.mean(y)
```

```python
np.median(y)
```

```python
x = scipy.stats.skewnorm.rvs(a=10, size=1000)
np.mean(x)
```

```python
np.median(x)
```

```python
sns.kdeplot(x)
plt.axvline(np.mean(x), color='red')
plt.axvline(np.median(x), color='green')
plt.show()
```

## Medidas de dispersión

### Rango

```python
np.min(x)
```

```python
np.max(x)
```

```python
np.max(x) - np.min(x)
```

### Desviación estándar

```python
x = np.random.normal(size=1000)
```

```python
np.mean(x)
```

```python
np.std(x)
```

```python
sns.kdeplot(x)
plt.axvline(np.std(x), color='red')
plt.axvline(-np.std(x), color='red')
plt.axvline(np.std(x)*2, color='green')
plt.axvline(-np.std(x)*2, color='green')
plt.axvline(np.std(x)*3, color='blue')
plt.axvline(-np.std(x)*3, color='blue')
plt.show()
```

```python
a = np.random.normal(scale=1, size=1000)
b = np.random.normal(scale=0.5, size=1000)
c = np.random.normal(scale=2, size=1000)
```

```python
sns.kdeplot(a)
sns.kdeplot(b)
sns.kdeplot(c)
plt.show()
```

### Varianza

```python
np.var(x)
```

### Rango intercuartilo

```python
np.percentile(x, 75) - np.percentile(x, 25)
```

```python
sns.boxplot(y=x)
plt.show()
```

# Descripción de variables categóricas

```python
x = ['F','M','F','M','F','M','F','M','F','M','F','M','F','F']
print(x)
```

```python
x = pd.DataFrame(x, columns=['sexo'])
x
```

```python
x.value_counts()
```

```python
x.value_counts(normalize=True)
```

```python
sns.countplot(x='sexo', data=x)
plt.show()
```
